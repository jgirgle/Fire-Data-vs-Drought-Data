```python
# importing dependencies
import pandas as pd
import pathlib

filepath = "drought_2014-2015.csv"

# reading drought CSV into dataframe
drought_df = pd.read_csv(filepath)
drought_df
```

```python
# finding unique values for the county column and passing it to its own dataframe 
county_df = pd.DataFrame(drought_df["County"].unique())
county_df
```

```python
# county_df["county_name"] = county_df[0]
county_df.reset_index()
```

```python
# creating unique id for each county from range 1 to 59, non-inclusive of 59
county_df["county_id"] = range(1,59)
```

```python
# passing county list and ids to a csv
county_df.to_csv("county_df.csv")
```
