```python
# importing dependencies
import pandas as pd
import pathlib
import datetime

filepath = "CA_Fires_2014-2015.csv"

# reading drought CSV into dataframe
fires_df = pd.read_csv(filepath)
fires_df
```

```python
fires_df.dtypes
```

```python
fires_df.columns
```

```python
fires_df.drop(columns=["DISCOVERY_DATE", "Unnamed: 0", "CONT_DATE", "CONT_DOY"], inplace=True)
```

```python
fires_df.rename(columns={"Unnamed: 0.1":"FIRE_ID"}, inplace=True)
fires_df
```

```python
fires_df["DISCOVERY_DATE"] = ""
fires_df
```

```python
# converting DOY to date
discovery_date = []

for x in range(0, len(fires_df)):
    YEAR = fires_df['FIRE_YEAR'][x]
    DAY_OF_YEAR = fires_df['DISCOVERY_DOY'][x]
    d = datetime.date(YEAR, 1, 1) + datetime.timedelta(int(DAY_OF_YEAR - 1))
    fires_df['DISCOVERY_DATE'][x] = d
```

```python
fires_df
```

```python
fires_df.drop(columns=["DISCOVERY_DOY", "DISCOVERY_TIME"], inplace=True)
```

```python
fires_df.to_csv("2014-15-CA-FIRES.csv")
```
