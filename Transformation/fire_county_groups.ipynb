```python
# importing dependencies
import pandas as pd
import pathlib
import datetime

filepath = "2014-15-CA-FIRES.csv"

# reading drought CSV into dataframe
fires_df = pd.read_csv(filepath)
fires_df
```

```python
fire_groups_df = fires_df.groupby(by=["FIRE_YEAR", "COUNTY_NAME"]).count()
fire_groups_df = pd.DataFrame(fire_groups_df["FIRE_ID"])
```

```python
fire_groups_df = fire_groups_df.reset_index()
fire_groups_df["NUM_OF_FIRES"] = fire_groups_df["FIRE_ID"]
```

```python
fire_groups_df.drop(columns=["level_0", "index", "FIRE_ID"], inplace=True)
```

```python
fire_groups_df.to_csv("fire_county_groups.csv")
```
