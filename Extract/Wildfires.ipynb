```python
# Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine
```

```python
# Path to sqlite database file
database_path = "data/FPA_FOD_20170508.sqlite"
```

```python
# Create Engine
engine = create_engine(f"sqlite:///{database_path}")
```

```python
engine.table_names()
```

```python
# Here we need to create a new connection object, which will serve the same purpose as the session
# the main distinction is the Session handles our (the user's) queries, the connection will handle queries from pandas
con = engine.connect()
```

```python
# Query All Records in the the Database
results_set_df = pd.read_sql("SELECT * FROM Fires Limit 10", con)
```
